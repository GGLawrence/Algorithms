Algorithm 01Knapsack(S, W):

    Input: S is a List of n Items, with each item's benefit bi, and each item's weight wi.     
           W is the maximum weight of this Knapsack.
           
    Output: The maximum benefit M[n, W] for this knapsack.
    
    // Initializaiton
    
    Create a n * W Metrix M
    
    for i <- 0 to n do
    
        for w <- 0 to W do
        
            M[i, j] <- 0            
    
    for i <- 1 to n do
    
        for w <- 1 to W do
        
            if w[i] <= w then
            
                M[i, w] <- bi + M[i - 1, w - w[i]]
                
            M[i, w] <- Max{M[i, w], M[i-1, w]}
            
    return M[n, W]

# The different between 0/1 Knapsack and Fractional Knapsack

## 0/1 Knapsack

We can not take fractional amounts from each item. Because, the weight of each item is fixed value, and we can not split them.

The goal: maximize $\sum_{i\in S}b_i$

Solution: Dynamic Programming

Dynamic programming is better than Greedy Algorithm for the "0/1 Knapsack" problem.

Reason:

DP can construct simple subproblems, create subproblem optimality, and solve the trouble of subproblem overlap. In this context, DP can globally explore all possible choices and consequences, ensuring a globally optimal solution.

While Greedy Algorithm seems faster, it may not guarantee global optimality, because it always does locally optimal choices without considering the global impact.

## Fractional Knapsack

We can take fractional amounts from each item. We can choose xi weight from item i with total weight wi, and xi <= wi.

The goal: maximize $\sum_{i\in S}(b_i * (x_i / w_i))$

Solution: Dynamic Programming

For fractional Knapsack problem, we do not think about the subproblem overlap. The Sum of local optimal is the global optimal. And Greedy algorithm is faster than Dynamic programming for this question.

## Algorithm

Algorithm TaskScheduling(T):

    Input: a set of Tasks T, with each task's start time si, and finish time fi.
    
    Output: A non-conflict Schedule of the Tasks in T, using minimal number of Machines m.
    
    m <- 0

        remove a task i from T with a smallest start time si
        
        if there is a Machine j can shedule task i without time conflicting
        
            schedule task i on Machine j
            
        else
        
            m <- m + 1

    return m
    
    
    
    
    
    
    



# MST

## Define

Given a weighted connnected undirected simple Graph G, the spanning tree of Graph G with minimal total edge weight is MST.

## Properties

### 1. Circle Properties

Given a weighted connected undirected simple Graph G, if an edge e is not in the MST named T, then e's weight will be at least greater as any edge in the circle created by adding e into T.

#### Proof by contradiction:
Assume the weight of edge e is smaller than one edge f in this circle, and f ∈ T. And then we can replace f with e in the T, and create a new spanning tree T'. Obviously, total weight(T') > total weight(T). It contradicts the fact T is MST. Therefore, edge e must be at least greater as any edge in the circle created by adding e into T.

### 2.  Partition Property

Given a weighted connected undirected simple Graph G, and for Graph G, given a partition of vertices of Graph G into 2 subsets U and V. Let e be an edge that connects U and V with minimal edge weight. Then the MST named T of this G, must have this edge e in it.

#### Proof by contradiction:

Assume there is an edge f connecting U and V, weight(f) > weight(e), f ∈ T, e ∉ T. We can create a new Spanning tree through replacing f with e in T. Thus, weight(T') < weight(T), and it contradicts the fact that T is MST. Therefore no such edge f exists.

### 3 Review proofs from assignment relating to trees, cycles, MST

Suppose G is an undirected, connected, weighted graph such that the edges in G have distinct edge weights. Show that the minimum spanning tree for G is unique.

#### Proof:
Assume given the edges in G have distinct edge weights, the minimum spanning tree for G is not unique. That is to say that there are at least 2 MSTs in G. Thus weight(T1) = weight(T2). There must be an edge e which is only in T1, but not in T2. Add e into T2 to create a circle. In this circle, we set f as the edge with the same node. Because the edges in G have distinct edge weights, thus weight(e) ≠ weight(f). If e > f, then we can create a new spanning tree T1', replacing e with f in T1. weight(T1'）< weight(T1) = weight(T2), it contradicts the fact T1 and T2 are MST. Therefore, the minimum spanning tree for G is unique.

# Algorithm

Algorithm KruskalMST(G):

    Input: a......Graph G with n vertices and m edges
    
    Output: MST stored in a set of edge T
    
    for each vertex v do
    
        C(v) <- {v}
    
    Create a PQ named Q to store all edges, using weight of edge as key
    
    T <- ∅
    
    while T has fewer than n - 1 edges do
    
        (u,v) <- Q.removeMin()
        
        Let C(u) be the cluster containing vertex u
        
        Let C(v) be the cluster containing vertex v
        
        if C(u) ≠ C(v) then 
        
            Union C(u) and C(v)
            
            add edge (u,v) into T
            
    return T
    
    

### Runtime analysis

The first for-loop runtime O(n)

Kruskal Algo uses PQ to store all edges with edge weight as key. It is a heap-based PQ. Runtime: log(m)

For the while loop:

Invoking Q.removeMin() once will take log(m) time.

Invoking Union(A, B) once will take log(n) time average.

In worst case, While-loop will take O(n) * O(log n) = O(n log n) time.

Therefore, KruskalAlgo runtime = O(n log n) + O(n) = O(n log n)

# Exchange argument vs lower bound argument

Exchange argument can be used to prove the optimality of an algorithm. The proof method is often constraction. Exchange argument method will always construct a new solution as seems better than original Algo. When we replace some parameters for this new solution, we find this new solution's performance is not better than original Algo. It contradicts this assument. Therefore, it proves the original algo is still optimal and its correctness.

Lower-Bound argument can be used to prove the Algo has some limitations. These limitation can ensure the original Algo is optimal under these limitations, like time conlict, limited space or number of Machines. We always use proof by contradiction and Lower-Bound argument to prove some original Algo is optimal in some limited conditions. If exceeds this limitation, Algo will be not effective.

# DFS

## Algorithm

Algo DFS(G, v):

    Input: a Graph G
    
    Output: traversing all vertices and edges of connected component of v
    
    Label v as explored vertex
    
    for each incident edge e of v do
    
        if e = (v, u) is unexplored then
        
            if u is unexplored then
            
                label e as discovery edge
                
                DFS(G, u)
                
            else
            
                label e as back edge

# Property 1

DFS will traverse all edges and vertice of CC of v in G.

Proof by contradiction:

Assume exist one unvisited vertex u. For the adjacent vertex j of u, when DFS labels j as explored vertex, then DFS must go to visit j's incident edge and adjacent vertex including u. And run the process DFS(G, u), then label u as explored. It contradicts the assument u is unvisited. Therefore no such vertex exists.

# Property 2

The discovery edges created by DFS can become a spanning tree of CC of v.

When we use DFS to visit a vertex u from a explored vertex v, only if u is unexplored, DFS just labels edge (v,u) as discovery. Thus, discovery edges will not create a circle in them. They will be a tree. Besides these discovery connects all vertices of CC of v. Therefore, discovery edges form a Spanning tree of CC of v.

# runtime analysis

O(n + m)

        S2 <- MergeSort(S2)
        
        S <- Merge(S1, S2)
        
    return S
    
Algo Merge(S1, S2):

    Input: S1, S2 are 2 arrays
    
    output: A sorted merged array S
    
    i <- 0
    
    j <- 0
    
    k <- 0
    
    create a empty array S, with size |S1| + |S2|
    
    while i <= |S1|-1 and j <= |S2| - 1 do
    
        if S1[i] <= S2[j] then
        
            S[k] <- S1[i]
            
            i++
            
            k++
        
        if S1[i] > S2[j] then
        
            S[k] <- S2[j]
            
            j++
            
            k++
    
    while i <= |S1| - 1 do
    
        S[k] <- S1[i]
            
            i++
            
            k++
        
    while j <= |S2| - 1 do
    
        S[k] <- S2[j]
            
            j++
            
            k++
        
    return S
