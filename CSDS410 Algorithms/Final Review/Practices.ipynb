Dynamic programming is better than Greedy Algorithm for the "0/1 Knapsack" problem.

Solution: Dynamic Programming

For fractional Knapsack problem, we do not think about the subproblem overlap. The Sum of local optimal is the global optimal. And Greedy algorithm is faster than Dynamic programming for this question.

## Define

Given a weighted connnected undirected simple Graph G, the spanning tree of Graph G with minimal total edge weight is MST.

## Properties

### 1. Circle Properties

Given a weighted connected undirected simple Graph G, if an edge e is not in the MST named T, then e's weight will be at least greater as any edge in the circle created by adding e into T.

#### Proof by contradiction:
Assume the weight of edge e is smaller than one edge f in this circle, and f ∈ T. And then we can replace f with e in the T, and create a new spanning tree T'. Obviously, total weight(T') > total weight(T). It contradicts the fact T is MST. Therefore, edge e must be at least greater as any edge in the circle created by adding e into T.

### 2.  Partition Property

Given a weighted connected undirected simple Graph G, and for Graph G, given a partition of vertices of Graph G into 2 subsets U and V. Let e be an edge that connects U and V with minimal edge weight. Then the MST named T of this G, must have this edge e in it.

#### Proof by contradiction:

Assume there is an edge f connecting U and V, weight(f) > weight(e), f ∈ T, e ∉ T. We can create a new Spanning tree through replacing f with e in T. Thus, weight(T') < weight(T), and it contradicts the fact that T is MST. Therefore no such edge f exists.

Algorithm KruskalMST(G):

    Input: a......Graph G with n vertices and m edges

    for each vertex v do

Algo DFS(G, v):

    Input: a Graph G

    Label v as explored vertex
    
    for each incident edge e of v do
    
        if e = (v, u) is unexplored then
        
            if u is unexplored then
            
                label e as discovery edge
                
                DFS(G, u)
                
            else
            
                label e as back edge

Algo Merge(S1, S2):

    Input: S1, S2 are 2 arrays
    
    output: A sorted merged array S
    
    i <- 0
    
    j <- 0
    
    k <- 0
    
    create a empty array S, with size |S1| + |S2|
    
    while i <= |S1|-1 and j <= |S2| - 1 do
    
        if S1[i] <= S2[j] then
        
            S[k] <- S1[i]
            
            i++
            
            k++
        
        if S1[i] > S2[j] then
        
            S[k] <- S2[j]
            
            j++
            
            k++
    
    while i <= |S1| - 1 do
    
        S[k] <- S1[i]
            
            i++
            
            k++
        
    while j <= |S2| - 1 do
    
        S[k] <- S2[j]
            
            j++
            
            k++
        
    return S
    
MergeSort uses extra space to store sorted array, so it is outplace

For the number in array with same value, for example S[i] = S[j], and i < j, after MergeSort, S[i'] = S[j'], i' is still less
